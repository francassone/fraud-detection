```python
# Function to calculate outliers using IQR method
def calculate_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    return len(outliers), lower_bound, upper_bound
```

```python
# Create box plots for all features
plt.figure(figsize=(20, 40))
for i, column in enumerate(df.columns[:-1]):  # Excluding the Class column
    plt.subplot(10, 3, i+1)
    sns.boxplot(x='Class', y=column, data=df)
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

```python
# Print outlier statistics for each feature
print("\nOutlier Analysis:")
print("-" * 50)
for column in df.columns[:-1]:  # Excluding the Class column
    # Calculate outliers for non-fraud transactions
    n_outliers_non_fraud, lb_non_fraud, ub_non_fraud = calculate_outliers(df[df['Class']==0], column)
    
    # Calculate outliers for fraud transactions
    n_outliers_fraud, lb_fraud, ub_fraud = calculate_outliers(df[df['Class']==1], column)
    
    print(f"\nFeature: {column}")
    print(f"Non-Fraud transactions outliers: {n_outliers_non_fraud}")
    print(f"Non-Fraud bounds: [{lb_non_fraud:.3f}, {ub_non_fraud:.3f}]")
    print(f"Fraud transactions outliers: {n_outliers_fraud}")
    print(f"Fraud bounds: [{lb_fraud:.3f}, {ub_fraud:.3f}]")
```
